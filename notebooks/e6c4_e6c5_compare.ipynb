```python
import sys
from pathlib import Path

# Make the scripts directory importable regardless of current working directory
scripts_dir = Path("scripts")
if not scripts_dir.exists():
    scripts_dir = Path("notebooks") / "scripts"

sys.path.insert(0, str(scripts_dir.resolve()))

from schema_compare import compare_schemas, DB_E6C4, DB_E6C5
```

```python
from IPython.display import display

diff_summary, table_details = compare_schemas(DB_E6C4, DB_E6C5, left_label="e6c4", right_label="e6c5")

if diff_summary.empty:
    print("Schemas match.")
else:
    print("Schema differences:")
    display(diff_summary)

    if table_details:
        print("\nDetailed table differences:")
        for table_name, detail_df in table_details.items():
            print(f"\n--- {table_name} ---")
            display(detail_df)
```
